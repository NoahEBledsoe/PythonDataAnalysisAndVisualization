```python
#Matplotlib is a very popular data visualization tool in Python 
#The most used component of Matplotlib is pyplot which is what we will be using
import numpy as np
import matplotlib.pyplot as plt
```

```python
#One of the most basic examples of pyplot is a line graph with a simple list
plt.plot([1,7,9,14])
#It assumes the numbers you pass are y-values and the width of the x-axis is how many numbers you pass
```

```python
#To specify your x-values, just pass it as the first parameter in plot()
plt.plot([1,2,7,8], [1,4,6,15])
```

```python
#The plot function also accepts a third parameter to signify the style of the line
#Check out the documentation online for a full list of modifications you can make but I will show how to make it
#red circles instead of a blue line
plt.plot([1,2,7,8], [1,4,6,15], 'ro')
```

```python
#Typically, we want to use numpy arrays instead of normal lists
myArray = np.array([[1,2,3,4],
                    [5,6,7,8]])

#To plot a 2d array, we'll say that index 0 is the x-axis and index 1 is the y-axis
plt.plot(myArray[0], myArray[1], 'go')
```

```python
#For a graph, we typically want to have labels for our axis' so people know what they're reading
plt.xlabel('the x axis')
plt.ylabel('the y axis')
plt.plot(myArray[0], myArray[1], 'b--')
```

```python
#We can also graph multiple sets of data on the same visual

#Creating a new array with the 'linspace' function in numpy to say we want an array with values from 0-5
#and 20 values inbetween
newArray = np.linspace(0,5,20)

#Showing some other arguments you can use to cusomize your visual (newArray**2 is the original array squared)
plt.plot(newArray, newArray**2, color='coral', linestyle='--')
plt.plot(newArray, 4*newArray, color='green', linewidth=2)
```

```python
#We can use the subplot function to change the number of rows and columns
#This example will have 1 row and 3 columns with an index of 1-3
plt.subplot(1,3,1)
plt.plot(newArray, newArray, 'c--')

plt.subplot(1,3,2)
plt.plot(newArray, newArray**2, 'b-')

plt.subplot(1,3,3)
plt.plot(newArray, newArray**3, 'mo')
```

```python
#Another very common graph in pyplot is the bar graph

#Let's say we want to compare grades of the people in the class
grades = {'Noah': 88,
          'Hari': 94,
          'Michael': 85,
          'Zach': 82,
          'Pralad': 79}

#The first argument is x-axis and the second argument is the y-axis
plt.bar(range(len(grades)), list(grades.values()))

#To change the name of the ticks on the x axis, just use the xticks function. We can also change the labels like before
plt.xticks(range(len(grades)), list(grades.keys()))
plt.xlabel('Student Name')
plt.ylabel('Final Grade')
```

```python
#Histograms are yet another very common graph

#Creating an arbitrary array of a random 1000 numbers to get a distribution
x = np.random.randn(1000)

#The first argument of the histogram is your data 
plt.hist(x, color='seagreen')
```
