```python
#Importing the libraries needed
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sea
```

```python
#Reading in our Pokemon dataset into a dataframe and printing out the top 5 rows
df = pd.read_csv(r'C:\Users\noahe\Documents\Pokemon.csv', index_col = 0, encoding = 'unicode-escape')
df.head()
```

```python
#Making a scatterplot with the lmplot() function which takes a value for the x-axis and y-axis and the data
sea.lmplot(x = 'Attack', y = 'Defense', data = df)
```

```python
#The lmplot function also takes in other arguments that allow you to encode even more information in your visual
sea.lmplot(x = 'Attack', y = 'Defense', data = df, fit_reg = False, hue = 'Stage')
```

```python
#Another important visual for looking at distributions is a boxplot
#We'll make a copy of the dataframe but exlucde a few attributes we don't care about and then show the graph
df_copy = df.drop(['Total', 'Stage', 'Legendary'], axis = 1)
sea.boxplot(data = df_copy)
```

```python
#Another very similar graph to a boxplot is the violinplot which has a "thickness" quality to show where the majority
#of the values are
sea.violinplot(data = df_copy)
```

```python
#We can also get an idea of how a Pokemon's type corresponds with their type (grass, water, poison, etc) with a violin
sea.violinplot(x = 'Type 1', y = 'Attack', data = df_copy)

#You can see the graph looks a bit crammed, we can resize the graph with the figure() function
plt.figure(figsize = (10,6))
sea.violinplot(x = 'Type 1', y = 'Attack', data = df_copy)
```

```python
#An important graph for looking at correlations between the data is a heatmap.
#To create a heatmap, we first need to use the corr() function to get a correlation
corr = df_copy.corr()
sea.heatmap(corr)
```

```python
#Histograms are also available and are great for showing distributions of numerical data
#Let's say we want an idea of the general distribution of the attack values
sea.distplot(df.Attack, color = 'green')
```

```python
#Count plots are great for seeing disributions of categorical data
#For example, let's say we want to see a count of each type of pokemon (grass, fire, etc)
sea.countplot(x = 'Type 1', data = df)

#We can see that the x axis is a bit crammed again so let's rotate the labels to be at a -45 degree angle
plt.xticks(rotation = -45)
```

```python
#The last plot we'll look at is a density plot. It's good for looking at distributions between two variables
#For example, let's compare the Attack and Defense attributes
sea.jointplot(df.Attack, df.Defense, kind = 'kde', color = 'purple') #kde is the keyword for density plots
```
